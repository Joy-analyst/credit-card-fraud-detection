```python
from scipy import stats

# 2. Generate Normal Data
normal_data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10

# 3. Calculate Mean and Median
mean_normal = np.mean(normal_data)
median_normal = np.median(normal_data)

print(f"Mean of normal data: {mean_normal:.2f}")
print(f"Median of normal data: {median_normal:.2f}")

# 4. Plot Normal Distribution
plt.figure(figsize=(8, 5))
sns.histplot(normal_data, kde=True, color='skyblue', bins=30)
plt.axvline(mean_normal, color='red', linestyle='--', label='Mean')
plt.axvline(median_normal, color='green', linestyle='-', label='Median')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# 5. Create Skewed Data
skewed_data = stats.skewnorm.rvs(a=10, loc=50, scale=10, size=1000)  # Positively skewed

# 6. Summary of Skewed Data
print("\nSkewed Data Summary:")
print(f"Mean: {np.mean(skewed_data):.2f}")
print(f"Median: {np.median(skewed_data):.2f}")
print(f"Skewness: {stats.skew(skewed_data):.2f}")

# 7. Plot Skewed Distribution
plt.figure(figsize=(8, 5))
sns.histplot(skewed_data, kde=True, color='salmon', bins=30)
plt.title('Positively Skewed Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Show missing values count
print("\nMissing Values in Data:")
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Step 1: Define X and y
X = df.drop('Class', axis=1)
y = df['Class']

# Step 2: Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # This gives you a NumPy array

# (Optional) Convert to DataFrame if you want to keep column names
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# Step 3: Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
print("Training Set Shape:", X_train.shape)
print("Testing Set Shape:", X_test.shape)
```

```python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print ("y_train shape:", y_train.shape)
print ("y_test shape:", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Train the model

model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
#predict on test set
y_pred = model.predict(X_test)
```

```python
#Evaluate performance 

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nclassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# After model.predict()
y_pred = model.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, model.predict_proba(X_test)[:,1]))
```
